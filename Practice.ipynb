### More Advanced question for visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Top ten companies which have launched most models ?
top_companies = mobile['Company Name'].value_counts().head(10)  # Corrected the syntax error by closing the head() method properly
print(top_companies)
```

```python
sns.barplot(x=top_companies.index, y=top_companies.values)
plt.xticks(rotation=90)
plt.xlabel('Company Name')
plt.ylabel('Number of Models')
plt.title('Top 10 Companies with Most Models')
plt.show()
```

```python
# What is the distribution of mobile prices in India ?
sns.boxplot(x='Launched Price (India)', data=mobile)
plt.title('Distribution of Mobile Prices in India')
plt.xlabel('Price in India')
plt.show()
```

```python
# 2. What is the distribution of mobile prices in Pakistan ?
sns.boxplot(x='Launched Price (Pakistan)', data=mobile)
plt.title('Distribution of Mobile Prices in Pakistan') 
plt.xlabel('Price in Pakistan')
plt.show()
```

```python
# How many mobiles are launched each year ?
mobile['Launched Year'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Mobiles Launched Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Mobiles')
plt.show()
```

```python
# How does RAM vary by brand ?
sns.boxplot(x='Company Name', y='RAM', data=mobile)
plt.title('RAM Variation by Brand') 
plt.xlabel('Company Name')
plt.ylabel('RAM (GB)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Is there a trend between battery capacity and phone weight?
sns.scatterplot(x='Battery Capacity', y='Mobile Weight', data=mobile)
plt.title('Battery Capacity vs Mobile Weight')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Mobile Weight (kg)')
plt.show()
```

```python
# top 10 processor types are most common across the dataset?
processor_counts = mobile['Processor'].value_counts().head(10)  
print(processor_counts)


sns.barplot(x=processor_counts.index, y=processor_counts.values, palette='viridis')
plt.title('Top 10 Processor Types')
plt.xlabel('Processor Name')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# What are the average prices of mobiles in USA ?
# if USA price is string then convert to float
mobile['Launched Price (USA)'] = mobile['Launched Price (USA)'].str.extract('(\d+\.?\d*)').astype(float)
# calculate average price
average_price_usa = mobile['Launched Price (USA)'].mean()
```

```python
print(f"The average price of mobiles in USA is: {average_price_usa} USD")
```

```python
sns.barplot(x='Company Name', y='Launched Price (USA)', data=mobile, estimator='mean')
plt.title('Average Mobile Prices in USA by Company')
plt.xlabel('Company Name')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Does the screen size affect the price of the mobile?

mobile['Screen Size'] = mobile['Screen Size'].str.extract('(\d+\.?\d*)').astype(float)
mobile['Launched Price (India)'] = mobile['Launched Price (India)'].str.extract('(\d+\.?\d*)').astype(float)


sns.scatterplot(x='Screen Size', y='Launched Price (India)', data=mobile)
plt.title('Screen Size vs Price in India')
plt.xlabel('Screen Size (inches)')
plt.ylabel('Price in India (INR)')
plt.show()
```

```python
# Which brands offer phones with the best battery-to-weight ratio?

mobile['Battery Capacity'] = mobile['Battery Capacity'].str.extract('(\d+\.?\d*)').astype(float)
mobile['Mobile Weight'] = mobile['Mobile Weight'].str.extract('(\d+\.?\d*)').astype(float)


mobile['Battery-to-Weight Ratio'] = mobile['Battery Capacity'] / mobile['Mobile Weight']
battery_weight_ratio = mobile[['Company Name', 'Model Name', 'Battery-to-Weight Ratio']].sort_values(by='Battery-to-Weight Ratio', ascending=False).head(10)
print(battery_weight_ratio)
```

```python
# top 5 brands with the best battery-to-weight ratio
top_brands_battery_weight = mobile.groupby('Company Name')['Battery-to-Weight Ratio'].mean().sort_values(ascending=False).head(5)
print(top_brands_battery_weight)
```

```python
# top brands barplot
sns.barplot(x=top_brands_battery_weight.index, y=top_brands_battery_weight.values, palette='viridis')
plt.title('Top 5 Brands with Best Battery-to-Weight Ratio')
plt.xlabel('Company Name')
plt.ylabel('Average Battery-to-Weight Ratio')
plt.xticks(rotation=90)
plt.show()
```
